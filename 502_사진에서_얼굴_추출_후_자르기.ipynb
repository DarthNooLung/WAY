```python
import cv2
```

```python
face_cascades = cv2.CascadeClassifier('./98_LearningData/00_haarcascades/haarcascade_frontalface_default.xml')
eye_cascades = cv2.CascadeClassifier('./98_LearningData/00_haarcascades/haarcascade_lefteye_2splits.xml')
```

```python
def trim_face(img, fileName):
        #이미지를 흑백으로 변경
        grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        #얼굴 검출
        faces = face_cascades.detectMultiScale(grayscale_img, scaleFactor=1.1, 
                                               minNeighbors=5, 
                                               minSize=(100,100), 
                                               flags=cv2.CASCADE_SCALE_IMAGE)
        
        iOrd = 1        
        
        #검출한 얼굴 주위에 테두리를 그린다
        for(x, y, width, height) in faces:
            #이미지 자르기
            img_trim = img[y:y+height, x:x+width]
                        
            eye_list = eye_cascades.detectMultiScale(img_trim,scaleFactor = 1.15, minNeighbors=1,minSize=(10,10))
            
            #눈이 있을 경우(잡힌게 2개 이상)
            if len(eye_list) >= 2:
                '''눈 부분에 표시 처리
                for eye in eye_list:
                    eyeX,eyeY,eyeW,eyeH = eye
                    cv2.rectangle(img_trim, (eyeX, eyeY), (eyeX + eyeW, eyeY + eyeH), (0, 255, 0), thickness=8)
                '''
                #파일명 만들기
                timFileName = fileName.replace(".jpg", "")
                timFileName += "_" + str(iOrd) + ".jpg"            
                
                #이미지 사이즈 리사이즈 하기
                img_trim = cv2.resize(img_trim,(200,200))
                
                #이미지 흑백 처리
                img_trim = cv2.cvtColor(img_trim, cv2.COLOR_BGR2GRAY)

                #자른 이미지 저장
                cv2.imwrite('./98_LearningData/03_Trim/' + timFileName, img_trim)

                print(str(iOrd) + "번째 얼굴 검출 완료[눈 : " + str(len(eye_list)) + "]")
                iOrd += 1
```

```python
import os
```

```python
files = os.listdir('./98_LearningData/01_Origin')
images = [file for file in files if 'jpg' in file]
```

```python
for image in images:
        print("=============================[" + image + "] 얼굴 검출 시작=============================")
        img = cv2.imread('./98_LearningData/01_Origin/' + image)
        trim_face(img, image)
        print("=============================[" + image + "] 얼굴 검출 종료=============================")
```
